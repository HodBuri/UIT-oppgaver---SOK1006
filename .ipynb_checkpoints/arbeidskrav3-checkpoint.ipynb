\begin{equation}
   x^{D}= 500 -3.2p\tag{1}
\end{equation}

og tilbud:


\begin{equation}
   x^{S}= -100 + 4.3(p-t)\tag{2}
\end{equation}

1. Finn likevektspris og -kvantum for alle verdier av $t$.        
2. Finn uttrykk for konsument- og produsentoverskudd, samt avgiftproveny.           
2. Tegn likevekten for $t=0$ og fargelegg KO og PO.     
2. Tegn inn likevekten for $t=4$, og fargelegg avgiftproveny.

Createing demand and supply function

```python

def demand(p:float=0.0,t:float=0.0):
    return 500-3.2*p

def supply(p:float=0.0,t:float=0.0):
    return -100+4.3*(p-t)
```

Basic Functions

```python
# equilibrium price and quantity function
def equilib(_supply,_demand,extra_change_demand:float = 0,extra_change_supply:float = 0):
    x,p,t,T=sp.symbols('x p t T', positive=True, real=True)
    eq_cond=sp.Eq(_demand(p,t),_supply(p,t))
    
    y_eq=sp.solve(eq_cond,p)
    y_eq_fn=sp.lambdify((t,T), y_eq[0])
    
    x_eq=_supply(y_eq[0],t)
    x_eq_fn=sp.lambdify((t,T), x_eq)
    
    y = y_eq_fn(extra_change_supply,extra_change_demand)
    x = x_eq_fn(extra_change_supply,extra_change_demand)
    return x,y

"""
def equilib2(_supply,_demand):
    x=sp.symbols('x', real=True, positive=True)
    equ=sp.Eq(_demand(x),_supply(x))
    x_eq=sp.solve(equ,x)
    x_eq=float(x_eq[0])
    p_eq=demand(x_eq) 
    return x_eq,p_eq
    """

# surpluss function
def consumer_surplus():
    pass

# draw lines to x and y ax
def graph_xy_line(ax,x,y):
    return ax.plot([0,x], [y, y], linestyle='dashed', color='white'),ax.plot([x,x], [y, 0], linestyle='dashed', color='white')



# create graph function
def create_ax(x_label:str='x_label',y_label:str='y_label',lim:list=[(0,None),(0,None)],x_pos:str = 'right',y_pos:str='top'):
                fig, ax = plt.subplots(figsize=(10,5))
                ax.set_ylabel(y_label, loc=y_pos)
                ax.set_xlabel(x_label, loc=x_pos)
                ax.set(xlim=lim[0])
                ax.set(ylim=lim[1])
                ax.spines['top'].set_color('none')
                ax.spines['right'].set_color('none')
                
                return fig, ax
```

```python
Ex, Ey = equilib(supply,demand)

x=sp.symbols('x', real=True, positive=True)
producer_surplus=sp.integrate(Ey-supply(x),(x,0,Ex))

consumer_surplus=sp.integrate(demand(x)-Ey,(x,0,Ex))

welfare_surplus=float(sp.integrate(demand(x)-supply(x),(x,0,Ex)))
welfare_surplus

print(producer_surplus,consumer_surplus,welfare_surplus)
```

Draw Graph

```python
graph_Xmin = 180
graph_Xmax = 300
graph_Ymin = 45
graph_Ymax = 105

Ex, Ey = equilib(supply,demand)
Ex2, Ey2 = equilib(supply,demand, extra_change_supply=4)

x_num=np.linspace(0,300,100)
q_num=np.linspace(0,Ex,100)
p_num=np.linspace(0,Ey,100)


fig, ax = create_ax(x_label='Ton per week',
                    y_label='Price per kg',
                    lim=[(graph_Xmin,graph_Xmax),(graph_Ymin,graph_Ymax)]
)


ax.plot(demand(x_num, 0),x_num, label='Demand')
ax.plot(supply(x_num, 0),x_num, label='Supply (t=0)')
ax.plot(supply(x_num, 4),x_num, label='Supply (t=4)')
ax.plot([0,Ex2], [Ey2-4, Ey2-4], linestyle='dashed', color='white')

ax.scatter(Ex,Ey,zorder=2, color='white')
ax.scatter(Ex2,Ey2,zorder=2, color='white')
ax.scatter(Ex2,Ey2-4,zorder=2, color='white')

graph_xy_line(ax,Ex,Ey)
graph_xy_line(ax,Ex2,Ey2)

ax.fill_between(supply(q_num),q_num,Ey,where=supply(q_num, 0)<=Ex,alpha = 0.3,label="Produsentoverskudd")#Produsentoverskudd
ax.fill_between(demand(q_num),q_num,Ey,where=demand(q_num, 0)<=Ex,alpha = 0.3,label="Konsumentoverskudd")#Konsumentoverskudd

ax.fill_between(q_num,Ey2,Ey2-4,where=q_num<=Ex2,color="#B06AFF",alpha = 0.5,label="Tax Revenu")#Tax
#ax.fill_between(supply(q_num, 0),q_num,demand(q_num,0),where=p_num<=Ex,color="black",alpha = 0.5,label="Death")#Death
#q3,supply(q3, 0),demand(q3),
#ax.fill_between(demand(q_num),q_num,Ey,where=p_num>=Ex,color="black",alpha = 1,label="Death")#Death
ax.fill_between(supply(q_num),q_num,Ey,where=q_num<=Ex,color="#FF21E1",alpha = 0.5,label="Death")#Death


#awnser_text = """ """
#ax.text(0,15, awnser_text, fontsize=8)

ax.text(Ex,graph_Ymin+1, f"{round(Ex,2)}", fontsize=10,rotation=-90)
ax.text(Ex2,graph_Ymin+1, f"{round(Ex2,2)}", fontsize=10,rotation=-90)

ax.text(graph_Xmin,Ey-4, f"{round(Ey,2)}", fontsize=10)
ax.text(graph_Xmin,Ey2+1, f"{round(Ey2,2)}", fontsize=10)


ax.set_title('Supply and Demand')
ax.legend(loc='best',frameon=False);
```
