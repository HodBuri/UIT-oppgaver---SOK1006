```python

def create_ax(x_label:str='x_label',y_label:str='y_label',lim:list=[(0,None),(0,None)],x_pos:str = 'right',y_pos:str='top'):
                fig, ax = plt.subplots(figsize=(10,5))
                ax.set_ylabel(y_label, loc=y_pos)
                ax.set_xlabel(x_label, loc=x_pos)
                ax.set(xlim=lim[0])
                ax.set(ylim=lim[1])
                ax.spines['top'].set_color('none')
                ax.spines['right'].set_color('none')
                
                return fig, ax

# Note to self, combinde to one function     
def create_ax2():
                fig, axs = plt.subplots(1,2, sharey=True, figsize=(15,5))
                for i in range(2):
                    axs[i].set_ylabel('Pris (p)', loc='top')
                    axs[i].spines['top'].set_color('none')
                    axs[i].spines['right'].set_color('none')
                return fig, axs
```

```python
def supply(x):
    return -2.0+1.7*x
    
def demand(x):
    return 8.5-0.82*x

def supply_new(x):
    return -2.0+1.7*x+0.4
```

```python
df_supply_demand = pd.DataFrame()
df_supply_demand['Q'] = list(range(20, 101))[0::5]
df_supply_demand['Log Q'] = np.log(df_supply_demand['Q'])
df_supply_demand['Supply (log P)'] = supply(df_supply_demand['Log Q'])
df_supply_demand['Demand (log P)'] = demand(df_supply_demand['Log Q'])
df_supply_demand['Supply (P)'] = np.exp(df_supply_demand['Supply (log P)'])
df_supply_demand['Demand (P)'] = np.exp(df_supply_demand['Demand (log P)'])
df_supply_demand['New Supply (log P)'] = supply_new(df_supply_demand['Log Q'])
df_supply_demand['New Supply (P)'] = np.exp(df_supply_demand['New Supply (log P)'])

df_supply_demand
```

```python
max_point = max(df_supply_demand['Demand (P)']) if max(df_supply_demand['Demand (P)']) >= max(df_supply_demand['Supply (P)']) else max(df_supply_demand['Supply (P)']) 
fig, ax = create_ax(x_label='Quantity (Q)',
                    y_label='Price (p)',
                    lim = [(min(df_supply_demand['Q']),max(df_supply_demand['Q'])),(0,max_point)]
)

q = sp.Symbol('q')
eq = sp.Eq(supply(q), demand(q))
max_q = sp.exp(sp.solve(eq)[0])
max_p = sp.exp(demand(sp.solve(eq)[0]))

ax.plot(df_supply_demand['Q'],df_supply_demand['Supply (P)'],label='Supply')
ax.plot(df_supply_demand['Q'],df_supply_demand['Demand (P)'],label='Demand')
ax.scatter(max_q,max_p,zorder=2);
ax.text(max_q-4,max_p-60, f'Price:{round(max_p,2)}\nQuantity:{round(max_q,2)}', fontsize=10)

ax.set_title('Supply and Demand')
ax.legend(loc='best',frameon=False);
```

```python
max_point = max(df_supply_demand['Demand (P)']) if max(df_supply_demand['Demand (P)']) >= max(df_supply_demand['Supply (P)']) else max(df_supply_demand['Supply (P)']) 
fig, ax = create_ax(x_label='Quantity (Q)',
                    y_label='Price (p)',
                    lim = [(min(df_supply_demand['Q']),max(df_supply_demand['Q'])),(0,max_point)]
)

q ,lq, lp = sp.symbols("q, lq, lp")
eq = sp.Eq(supply(q), demand(q))
max_q = sp.exp(sp.solve(eq)[0])
max_p = sp.exp(demand(sp.solve(eq)[0]))

new_eq = sp.Eq(supply_new(q), demand(q))
new_max_q = sp.exp(sp.solve(new_eq)[0])
new_max_p = sp.exp(demand(sp.solve(new_eq)[0]))

eq_supply = sp.solve(sp.Eq(supply(lq), lp),lq)
eq_demand = sp.solve(sp.Eq(demand(lq), lp),lq)

ax.plot(df_supply_demand['Q'],df_supply_demand['Supply (P)'],label='Supply')
ax.plot(df_supply_demand['Q'],df_supply_demand['Demand (P)'],label='Demand')
ax.plot(df_supply_demand['Q'],df_supply_demand['New Supply (P)'],label='New Supply')

ax.plot ([max_q,max_q], [max_p, 0], linestyle='dashed', color='white')
ax.plot ([0,max_q], [max_p, max_p], linestyle='dashed', color='white')
ax.plot ([new_max_q,new_max_q], [new_max_p, 0], linestyle='dashed', color='white')
ax.plot ([0,new_max_q], [new_max_p, new_max_p], linestyle='dashed', color='white')

ax.scatter(max_q,max_p,zorder=2);
ax.scatter(new_max_q,new_max_p,zorder=2);
ax.text(max_q-4,max_p+25, f'Price:{round(max_p,2)}\nQuantity:{round(max_q,2)}', fontsize=8)
ax.text(new_max_q-5,new_max_p+30, f'Price:{round(new_max_p,2)}\nQuantity:{round(new_max_q,2)}', fontsize=8)
ax.text(20,-50, f'Supply Elasticity : {round(0.588,2)}% change per 1% change in price (low elasticity)\nDemand Elasticity : {round(-1.219,2)}% change per 1% change in price (high elasticity)', fontsize=8)

ax.set_title('Supply Change')
ax.legend(loc='best',frameon=False);
#ax.legend(loc='lower center', bbox_to_anchor=(0.5, -.15),ncol=5,frameon=False)
```
